```python
import csv
import os
import random
os.chdir(r'C:\Users\shivani.ss\Documents\Assessment')
class Matrimony:
    def Menu(self):
        print("----------Welcome to our Matrimonial System----------")
        print("0 for Exit")
        print("1 for Insertion")
        print("2 for Modification")
        print("3 for Retrieval")
        print("4 for Searching")
        print("5 for Recommendation")
        print("6 for Deletion")
        try:
            self.choice=int(input("Enter Your Choice: "))
            return self.choice
        except ValueError:
            print("Enter Only Integer Choices")
    def Insertion(self):
        try:
            self.name=input("Enter Your Name: ")
            self.gender=input("Enter Your Gender: ")
            self.age=int(input("Enter Your Age: "))
            self.DOB=input("Enter Your DOB: ")
            self.Lang=input("Enter Language: ")
            self.Country=input("Your Country: ")
            self.state=input("Your State: ")
            self.Job=input("Job: ")
            self.Salary=input("Your Salary: ")
            self.Qual=input("Enter Qualification: ")
            self.uniq_id=input("Enter Your Contact number: ")
        except ValueError:
            print("Age Should be Integer")
        with open('DB.csv','a',newline="") as f:
            csvinsert=csv.writer(f)
            csvinsert.writerow([self.uniq_id,self.name,self.gender,self.age,self.DOB,self.Lang,self.Country,self.state,self.Job,self.Salary,self.Qual])
        print("----------Successfully Inserted----------")
    #def Modification(self):
        
    def Retrieval(self):
        print("-"*60)
        print("\t\tuniq_id \t\tName \t\tGender \t\tAge \t\tAge \t\tDOB \t\tLanguage \t\tCountry \t\tState \t\tJob \t\tSalary \t\tQualification ")
        with open('DB.csv','r') as f:
            csv_retrieval=csv.reader(f)
            for i in list(csv_retrieval):
                print("-"*60)
                print(i[0],'\t\t',i[1],'\t\t',i[2],'\t\t',i[3],'\t\t',i[4],'\t\t',i[5],'\t\t',i[6],'\t\t',i[7],'\t\t',i[8],'\t\t',i[9],'\t\t',i[10])
        print("-"*60)
    def Searching(self):
        self.search=input("Enter data to search: ")
        print("------------------Search Results------------------------")
        print("-"*60)
        print("\t\tuniq_id \t\tName \t\tGender \t\tAge \t\tAge \t\tDOB \t\tLanguage \t\tCountry \t\tState \t\tJob \t\tSalary \t\tQualification ")
        print("-"*60)
        with open('DB.csv','r') as f:
            csv_retrieval=csv.reader(f)
            dlist=list(csv_retrieval)
            sresult=[data for data in dlist if self.search in data]
            if len(sresult)>0:
                for i in list(sresult):
                    print(i[0],'\t\t',i[1],'\t\t',i[2],'\t\t',i[3],'\t\t',i[4],'\t\t',i[5],'\t\t',i[6],'\t\t',i[7],'\t\t',i[8],'\t\t',i[9],'\t\t',i[10])
            else:
                print("----------------No Search Results Found----------------")
        print("-"*60)
    def Deletion(self):
        deldata=input("Enter Person name to delete:")
        with open('DB.csv','r')as file:
            data=csv.reader(file)
            dellist=list(data)
            f_result=[name for name in dellist if deldata != name[1]]
        with open('DB.csv','w',newline='')as file:
            csvwriter=csv.writer(file)
            csvwriter.writerows(f_result)
            if(len(f_result) == len(dellist)):
                print("No result")
            else:
                print("updated list")
                self.Retrieval()
    def Recommendation(self):
        with open('DB.csv','r') as f:
            csv_retrieval=csv.reader(f)
            for i in list(csv_retrieval):
                for j in list(csv_retrieval):
                    if(i[7]==j[7]):
                        print(j)
            
d=Matrimony()
c=d.Menu()
while True:
    if(c==0):
        break
    elif(c==1):
        d.Insertion()
        break
    elif(c==2):
        d.Modification()
        break
    elif(c==3):
        d.Retrieval()
        break
    elif(c==4):
        d.Searching()
        break
    elif(c==5):
        d.Recommendation()
        break
    elif(c==6):
        d.Deletion()
        break
    else:
        pass
```
